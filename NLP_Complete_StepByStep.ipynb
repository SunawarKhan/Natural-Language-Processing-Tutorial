# Part of Speech(PoS) Tagging
#### The Quick Brown Fox Jumps Over the Lazy Dog

```python
import nltk
from nltk.tokenize import word_tokenize

# Sample sentence
sentence = "The quick brown fox jumps over the lazy dog."

# Tokenize the sentence into words
words = word_tokenize(sentence)

# Perform POS tagging
pos_tags = nltk.pos_tag(words)

# Define a mapping dictionary for POS tags to descriptive names
pos_mapping = {
    'CC': 'Coordinating conjunction',
    'CD': 'Cardinal number',
    'DT': 'Determiner',
    'EX': 'Existential there',
    'FW': 'Foreign word',
    'IN': 'Preposition or subordinating conjunction',
    'JJ': 'Adjective',
    'JJR': 'Adjective, comparative',
    'JJS': 'Adjective, superlative',
    'LS': 'List item marker',
    'MD': 'Modal',
    'NN': 'Noun, singular or mass',
    'NNS': 'Noun, plural',
    'NNP': 'Proper noun, singular',
    'NNPS': 'Proper noun, plural',
    'PDT': 'Predeterminer',
    'POS': 'Possessive ending',
    'PRP': 'Personal pronoun',
    'PRP$': 'Possessive pronoun',
    'RB': 'Adverb',
    'RBR': 'Adverb, comparative',
    'RBS': 'Adverb, superlative',
    'RP': 'Particle',
    'SYM': 'Symbol',
    'TO': 'to',
    'UH': 'Interjection',
    'VB': 'Verb, base form',
    'VBD': 'Verb, past tense',
    'VBG': 'Verb, gerund or present participle',
    'VBN': 'Verb, past participle',
    'VBP': 'Verb, non-3rd person singular present',
    'VBZ': 'Verb, 3rd person singular present',
    'WDT': 'Wh-determiner',
    'WP': 'Wh-pronoun',
    'WP$': 'Possessive wh-pronoun',
    'WRB': 'Wh-adverb',
    '.': 'Punctuation'
}

# Print the POS tags along with their descriptive names
print("POS tags with their descriptive names:")
for word, pos_tag in pos_tags:
    print(f"{word}: {pos_tag} ({pos_mapping.get(pos_tag, 'Unknown')})")
```

### Hypernyms

```python
from nltk.corpus import wordnet

word = wordnet.synsets("Play")[0]

# Find the More Abstract Term
print(word.hypernyms())
```

```python
from nltk.corpus import wordnet

# Word Meaning with Definitions
for synset in wordnet.synsets("Fun"):
    print(synset.name())
    print(synset.definition())
    print(synset.examples())

    for lemma in synset.lemmas():
        print(lemma.name())
    print("\n")
```

### Hyponyms

```python
from nltk.corpus import wordnet

word = wordnet.synsets("Play")[0]

# Find the More Abstract Term
print(word.hyponyms())
```

### Synonyms

```python
from nltk.corpus import wordnet

#Finding Synonyms, Empty List to Store Synonyms
synonyms=[]

for words in wordnet.synsets("Fun"):
    for lemma in words.lemmas():
        synonyms.append(lemma.name())

synonyms
```

### Antonyms

```python
from nltk.corpus import wordnet

#Finding Synonyms, Empty List to Store Synonyms
antonyms=[]

for words in wordnet.synsets("Natural"):
    for lemma in words.lemmas():
        if lemma.antonyms():
            antonyms.append(lemma.antonyms()[0].name())

antonyms
```

```python
from nltk.corpus import wordnet
#Finding Synonyms, Empty List to Store Synonyms
synonyms=[]
#Finding Antonyms, Empty List to Store Antonyms
antonyms=[]
for words in wordnet.synsets("Now"):
    for lemma in words.lemmas():
        synonyms.append(lemma.name())
        if lemma.antonyms():
            antonyms.append(lemma.antonyms()[0].name())

print(synonyms)
print("\n")
print(antonyms)
```

## Preprocessing

```python
# Selecting the text and target column for further analysis
data = df[['text', 'target']]

# Replacing the values to ease understanding
data.loc[data['target'] == 4, 'target'] = 1

# Printing unique values of the target variable
print(data['target'].unique())
```

```python
#Separating positive and negative tweets
data_pos = data[data['target']==1]
data_neg = data[data['target']==0]
```

```python
#Combining positive and negative tweets
dataset = pd.concat([data_pos, data_neg])
```

```python
#Making statement text in lowercase
dataset['text']=dataset['text'].str.lower()
dataset['text'].tail()
```

### Stop Word

```python
#Defining set containing all stopwords in English.

stopwordlist = ['a', 'about', 'above', 'after', 'again', 'ain', 'all', 'am', 'an',
             'and','any','are', 'as', 'at', 'be', 'because', 'been', 'before',
             'being', 'below', 'between','both', 'by', 'can', 'd', 'did', 'do',
             'does', 'doing', 'down', 'during', 'each','few', 'for', 'from',
             'further', 'had', 'has', 'have', 'having', 'he', 'her', 'here',
             'hers', 'herself', 'him', 'himself', 'his', 'how', 'i', 'if', 'in',
             'into','is', 'it', 'its', 'itself', 'just', 'll', 'm', 'ma',
             'me', 'more', 'most','my', 'myself', 'now', 'o', 'of', 'on', 'once',
             'only', 'or', 'other', 'our', 'ours','ourselves', 'out', 'own', 're','s', 'same', 'she', "shes", 'should', "shouldve",'so', 'some', 'such',
             't', 'than', 'that', "thatll", 'the', 'their', 'theirs', 'them',
             'themselves', 'then', 'there', 'these', 'they', 'this', 'those',
             'through', 'to', 'too','under', 'until', 'up', 've', 'very', 'was',
             'we', 'were', 'what', 'when', 'where','which','while', 'who', 'whom',
             'why', 'will', 'with', 'won', 'y', 'you', "youd","youll", "youre",
             "youve", 'your', 'yours', 'yourself', 'yourselves']
```

### Removing Stop Words

```python
# Cleaning and removing the above stop words list from the tweet text
STOPWORDS = set(stopwordlist)
def cleaning_stopwords(text):
    return " ".join([word for word in str(text).split() if word not in STOPWORDS])
dataset['text']=dataset['text'].apply(lambda text: cleaning_stopwords(text))
dataset['text'].head()
```

### Removing Punctuations

```python
# Cleaning and removing punctuations
import string
english_punctuations = string.punctuation
punctuations_list = english_punctuations
def cleaning_punctuations(text):
    translator = str.maketrans('','',punctuations_list)
    return text.translate(translator)
dataset['text']=dataset['text'].apply(lambda x: cleaning_punctuations(x))
dataset['text'].tail()
```

### Removing Repeating Charactars

```python
# Cleaning and removing repeating characters
def cleaning_repeating_char(text):
    return re.sub(r'(.)1+', r'1',text)
dataset['text'] = dataset['text'].apply(lambda x: cleaning_repeating_char(x))
dataset['text'].tail()
```

### Removing URL's

```python
# Cleaning and removing URLs
def cleaning_URLs(data):
    return re.sub('((WWW.[^s]+)|(https?://[^s]+))',' ',data)
dataset['text']=dataset['text'].apply(lambda x: cleaning_URLs(x))
dataset['text'].tail()
```

### Removing Numerical Numbers

```python
# Cleaning and removing numeric numbers
def cleaning_numbers(data):
    return re.sub('[0-9]+', '', data)
dataset['text'] = dataset['text'].apply(lambda x: cleaning_numbers(x))
dataset['text'].tail()
```

### Tokenization

```python
#Getting Tokenization of TWEET Text
from nltk.tokenize import RegexpTokenizer

# Assuming 'text' column contains lists of words
dataset['text'] = dataset['text'].apply(lambda x: ' '.join(x) if isinstance(x, list) else x)
tokenizer = RegexpTokenizer(r'\w+')
dataset['text'] = dataset['text'].apply(tokenizer.tokenize)
dataset['text'].head()

# Assuming 'text' column contains lists of words
#dataset['text'] = dataset['text'].apply(lambda x: ' '.join(x) if isinstance(x, list) else x)
#dataset['text'] = dataset['text'].apply(tokenizer.tokenize)
#dataset['text'].head()
```

### Lemmatizer

```python
# Applying Lemmatizer
import nltk
lm = nltk.WordNetLemmatizer()
def lemmatizer_on_text(data):
    text = [lm.lemmatize(word) for word in data]
    return data
dataset['text'] = dataset['text'].apply(lambda x: lemmatizer_on_text(x))
dataset['text'].head()
```

### Stemming

```python
#Applying Stemming
import nltk
st = nltk.PorterStemmer()
def  stemming_on_text(data):
    text=[st.stem(word) for word in data]
    return data
dataset['text'] = dataset['text'].apply(lambda x: stemming_on_text(x))
dataset['text'].head()
```
